```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau=100e-6
E=4
```

```python
def impulse(t, E, tau):
    if abs(t) >  tau/2:
        return 0.0
    return E * np.cos(np.pi*t/tau)

def spectrum (f, E, tau):
    if np.isclose(f, 0):
        return E * tau/2
    return 0.5 * E * tau * ((np.sin(np.pi*(f-1/2/tau)*tau)/(np.pi*(f-1/2/tau)*tau))\
                       + (np.sin(np.pi*(f+1/2/tau)*tau)/(np.pi*(f+1/2/tau)*tau)))
```

```python
t_band = np.linspace(-tau, tau, num = 2**14)
f_band = np.linspace(-5/tau, 5/tau, num = 2**14)
x = np.array([impulse(t, E, tau) for t in t_band])
Xf = np.array([spectrum(t, E, tau) for t in f_band])
```

```python
plt.figure(figsize=(15, 7))
plt.plot(t_band*1e6, x)
plt.title("Импульс $x_0(t)$")
plt.xlabel("$t$, мкс")
plt.ylabel("$x_0(t)$, B")
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.title("АЧХ импульса $x_0(t)$")
plt.plot(f_band*1e-3, abs(Xf)*1e6)
plt.xlabel("$f$, кГц")
plt.ylabel("$|X_0|(f)$, мкB/Гц")
plt.tight_layout()
plt.grid()
plt.show()
```
